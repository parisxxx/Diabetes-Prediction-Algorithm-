```python
#splitting the dataset into features and target variable
feature_column = ['pregnant', 'insulin', 'bmi', 'age','glucose','bp','pedigree']
X = diabetes_data[feature_column]
y = diabetes_data.outcome
```

```python
#importing train test set method
from sklearn.model_selection import train_test_split
```

```python
#splitting the independent and dependent variables into training and testing sets and using random state method for same sets
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.28,random_state=0)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
# instantiating the model (using the default parameters)
logreg = LogisticRegression(max_iter = 1000)
#using the fit method to train the model
logreg.fit(X_train,y_train)
```

```python
#predic method on actual data set
y_pred=logreg.predict(X_test)
y_pred
```

```python
X_test
```

```python
# importing the metrics class to create confusion matrics
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
#Plotting the Confusion Matrics
class_names=[0,1] # name of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# creating a heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1,fontsize=30)
plt.ylabel('Actual outcome',fontsize=15)
plt.xlabel('Predicted outcome',fontsize=15)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
